```python
import pandas as pd
import numpy as np
```

```python
dados = {
    "nome": ["Lucas", "Mariana", "Pedro", "Juliana", "Rafael"],
    "idade": [22, 31, 27, np.nan, 45],
    "cidade": ["PE", "BA", "PE", "CE", "BA"],
    "salario": [2800, 5200, 4100, 3300, np.nan],
    "data_contratacao": pd.to_datetime([
        "2021-02-15", "2018-06-10", "2022-08-05", "2019-12-20", "2023-01-25"
    ])
}

df = pd.DataFrame(dados)
print("DataFrame inicial:\n", df)
```

```python
df.to_csv("dados.csv", index=False)
df = pd.read_csv("dados.csv")
```

```python
print("\nInfo:")
print(df.info())

print("\nEstatísticas:")
print(df.describe())

# =========================
# 4. Tratamento de dados
# =========================
df["idade"] = df["idade"].fillna(df["idade"].mean())
df["salario"] = df["salario"].fillna(df["salario"].median())
```

```python
df["idade"] = df["idade"].fillna(df["idade"].mean())
df["salario"] = df["salario"].fillna(df["salario"].median())
```

```python
df["salario_anual"] = df["salario"] * 12
```

```python
filtro = df[df["salario"] > 3500]
print("\nSalários maiores que 3500:\n", filtro)
```

```python
df_ordenado = df.sort_values(by="salario", ascending=False)
print("\nOrdenado por salário:\n", df_ordenado)
```

```python
media_por_cidade = df.groupby("cidade")["salario"].mean()
print("\nMédia salarial por cidade:\n", media_por_cidade)
```

```python
df["categoria_salario"] = df["salario"].apply(
    lambda x: "Alto" if x > 4500 else "Médio" if x > 3000 else "Baixo"
)
```

```python
pivot = pd.pivot_table(
    df,
    values="salario",
    index="cidade",
    columns="categoria_salario",
    aggfunc="mean"
)
print("\nPivot table:\n", pivot)
```

```python
bonus = pd.DataFrame({
    "nome": ["Lucas", "Mariana", "Pedro", "Juliana", "Rafael"],
    "bonus": [400, 1200, 800, 500, 900]
})

df_merge = pd.merge(df, bonus, on="nome")
print("\nDataFrame com bônus:\n", df_merge)
```

```python
df["ano_contratacao"] = pd.to_datetime(df["data_contratacao"]).dt.year
```

```python
df.to_excel("resultado.xlsx", index=False)

print("\nProcesso concluído!")
```
